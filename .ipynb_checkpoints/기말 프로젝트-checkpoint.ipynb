```python
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.keys import Keys
import time

wd = webdriver.Chrome() #크롬 webDriver 객체 생성
wd.get("https://pcmap.place.naver.com/place/list?query=잠실%20편의점&fromNxList=true&noredirect=1") #웹 페이지 연결

# 데이터수집
store_data = []

while True:
    # 스크롤을 반복적으로 내려 데이터 로드
    scroll_attempts = 5  # 스크롤 시도 횟수
    for _ in range(scroll_attempts):
        actions = wd.find_element(By.ID, '_pcmap_list_scroll_container')
        actions.send_keys(Keys.END)  # 페이지 끝으로 스크롤
        time.sleep(2)  # 데이터 로드 대기

    html = wd.page_source
    soup = BeautifulSoup(html, 'html.parser')

    # 현재 페이지의 li요소들 모두 가져옴
    stores = soup.find_all("li", class_="VLTHu OW9LQ")

    for store in stores:
        name = store.find("span", class_="YwYLL").text
        location = store.find("span", class_="Pb4bU").text
        store_data.append({"name": name, "location": location})
        
    #다음 페이지 버튼 클릭
    try:
        next_button = wd.find_element(By.XPATH, '//*[@id="app-root"]/div/div[2]/div[2]/a[6]')
        next_button.click()
        time.sleep(2)
    except:
        break
        

# 데이터 출력
for store in store_data:
    print(store)
```

```python
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.keys import Keys
import time
import pandas as pd

wd = webdriver.Chrome() #크롬 webDriver 객체 생성

incheon_locations = ["중구", "동구", "미추홀구", "연수구", "남동구", "부평구", "계양구", "서구", "강화군", "옹진군"]

incheon_stores_data = {
    "중구": 0, "동구": 0, "미추홀구": 0, "연수구": 0,
    "남동구": 0, "부평구": 0, "계양구": 0,
    "서구": 0, "강화군": 0, "옹진군": 0
}

#인천에 있는 지역 지도에 검색
for incheon_location in incheon_locations:
    wd.get(f"https://www.google.com/search?sca_esv=293021c43ebdc58d&tbm=lcl&q=인천+{incheon_location}+편의점&rflfq=1&num=10&sa=X&ved=2ahUKEwj_jKvcpY6KAxXMka8BHUmBB1AQjGp6BAgjEAE&biw=573&bih=632&dpr=1.38#rlfi=hd:;si:;")

    while True:
        html = wd.page_source
        soup = BeautifulSoup(html, 'html.parser')

        # 현재 페이지의 특정 클래스의 div요소들 모두 가져옴
        stores = soup.find_all("div", class_="VkpGBb")

        # 총 편의점 개수에 현재 페이지 편의점 수 저장
        incheon_stores_data[incheon_location] += int(len(stores))
        print(incheon_stores_data)

        try:
            # 다음 페이지 버튼 누름
            next_button = wd.find_element(By.XPATH, '//*[@id="pnnext"]/span[2]')
            next_button.click()
            time.sleep(2)
        # 다음페이지 버튼 못 누를 경우 반복문 멈춤
        except:
            break
        
#엑셀 파일로 저장
stores_result = pd.DataFrame(list(incheon_stores_data.items()), columns=['행정구', '편의점 개수'])
stores_result.to_csv('./인천광역시_행정구역별_편의점_개수.csv',index=False, encoding='utf-8')
print(incheon_stores_data)
```
